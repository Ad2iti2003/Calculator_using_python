```python
def calculate_avg(marks):
  return sum(marks)/len(marks)

student = {}
while True:
  print("1. Add Student")
  print("2. Show topper")
  print("3. Exit")

  choice = int(input("Enter choice"))

  if choice == 1:
    name = input("Enter your name")
    marks = list(map(int,input("enter your marks seperated by space").split()))
    student[name] = marks;

  elif choice == 2:
    topper = None
    highest_avg = 0

    for name,marks in student.items():
      avg = calculate_avg(marks)
      if avg > highest_avg:
        highest_avg = avg
        topper = name

    if topper:
      print("Topper:", topper, "Average:", highest_avg)
    else:
      print("No students added")

  else:
     break



```
